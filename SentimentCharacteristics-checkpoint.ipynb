```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from wordcloud import WordCloud
import re
```

```python
# Load the dataset (Replace 'FinalSentiment.csv' with the correct path if needed)
df = pd.read_csv('FinalSentiment.csv')

# Convert column names to lowercase for consistency
df.columns = df.columns.str.lower()

# Create a new column for text length
df['text_length'] = df['statement'].astype(str).apply(len)
```

```python
# Do Longer Texts Generally Have More Negative Sentiments?
plt.figure(figsize=(8,5))
sns.scatterplot(x='text_length', y='sentiment_score', data=df, alpha=0.5)
plt.title('Text Length vs. Sentiment Score')
plt.show()
print("Correlation coefficient:", df[['text_length', 'sentiment_score']].corr().iloc[0,1])
```

```python
# Most Common Words in Highly Negative Statements
from collections import Counter
negative_texts = " ".join(df[df['sentiment_score'] < -0.5]['statement'].dropna().astype(str))
negative_words = re.findall(r'\b\w+\b', negative_texts.lower())
word_freq = Counter(negative_words)
wordcloud = WordCloud(width=800, height=400, background_color='black').generate_from_frequencies(word_freq)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Highly Negative Statements')
plt.show()
```

```python
# Most Common Phrases in Positive Statements
positive_texts = df[df['sentiment_score'] > 0.5]['statement'].dropna().astype(str)
positive_phrases = Counter(" ".join(positive_texts).split()).most_common(20)
print("Most Common Phrases in Positive Statements:", positive_phrases)
```

```python
# Word Complexity (Average Word Length) vs Sentiment
df['avg_word_length'] = df['statement'].astype(str).apply(lambda x: np.mean([len(word) for word in x.split()]) if len(x.split()) > 0 else 0)
plt.figure(figsize=(6,4))
sns.scatterplot(x='avg_word_length', y='sentiment_score', data=df, alpha=0.5)
plt.title('Average Word Length vs. Sentiment Score')
plt.show()
print("Correlation coefficient:", df[['avg_word_length', 'sentiment_score']].corr().iloc[0,1])
```

```python
# Word Repetition and Sentiment
def count_repeated_words(text):
    words = text.lower().split()
    return sum(words.count(word) > 1 for word in set(words))
df['word_repetition'] = df['statement'].astype(str).apply(count_repeated_words)
plt.figure(figsize=(6,4))
sns.scatterplot(x='word_repetition', y='sentiment_score', data=df, alpha=0.5)
plt.title('Word Repetition vs. Sentiment Score')
plt.show()
print("Correlation coefficient:", df[['word_repetition', 'sentiment_score']].corr().iloc[0,1])
```
