```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Load dataset
df = pd.read_csv("FinalSentiment.csv")

# Convert 'date_collected' to datetime (if applicable)
df['date_collected'] = pd.to_datetime(df['date_collected'], dayfirst=True)
```

```python
# How does sentiment vary across different mental illness types?
plt.figure(figsize=(15,5))
sns.boxplot(x='mental_illness_type', y='encoded_status', data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Sentiment Variation Across Mental Illness Types")
plt.show()
```

```python
# Are there specific topics frequently discussed by users with different mental illnesses?
vectorizer = CountVectorizer(stop_words='english', max_features=20)
word_counts = pd.DataFrame(vectorizer.fit_transform(df['statement'].dropna()).toarray(), columns=vectorizer.get_feature_names_out())
top_words = word_counts.sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=top_words.index, y=top_words.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Most Common Words Used in Statements")
plt.show()
```

```python
# How does sentiment change before and after mentioning a mental illness?
df['before_after_mention'] = df.groupby('Mental_Illness_Category')['encoded_status'].diff()
plt.figure(figsize=(10,5))
sns.histplot(df['before_after_mention'].dropna(), bins=30, kde=True, color="purple")
plt.title("Sentiment Change Before and After Mental Illness Mention")
plt.show()
```

```python
# Are individuals who talk about therapy or medication more positive in sentiment?
# Assuming "therapy" or "medication" is mentioned in 'statement'
df['therapy_mentioned'] = df['statement'].str.contains("therapy|medication", case=False, na=False)
therapy_sentiment = df.groupby('therapy_mentioned')['encoded_status'].mean()
plt.figure(figsize=(6,4))
sns.barplot(x=therapy_sentiment.index, y=therapy_sentiment.values, palette="Set2")
plt.xticks([0, 1], ["No Therapy Mention", "Therapy Mentioned"])
plt.title("Sentiment Change with Therapy Discussion")
plt.show()
```

```python
# How often do individuals with different mental health issues use exclamatory sentences?
df['exclam_count'] = df['statement'].str.count("!")
exclam_usage = df.groupby('Mental_Illness_Category')['exclam_count'].mean()
plt.figure(figsize=(10,5))
sns.barplot(x=exclam_usage.index, y=exclam_usage.values, palette="flare")
plt.xticks(rotation=45)
plt.title("Average Use of Exclamation Marks by Mental Illness Category")
plt.show()
```

```python
# How does sentiment correlate with mentions of suicidal thoughts?
df['suicidal_mention'] = df['statement'].str.contains("suicide|kill myself|end my life", case=False, na=False)
suicide_sentiment = df.groupby('suicidal_mention')['encoded_status'].mean()
plt.figure(figsize=(6,4))
sns.barplot(x=suicide_sentiment.index, y=suicide_sentiment.values, palette="coolwarm")
plt.xticks([0, 1], ["No Mention", "Suicidal Mention"])
plt.title("Sentiment Scores in Statements with Suicidal Mentions")
plt.show()
```
